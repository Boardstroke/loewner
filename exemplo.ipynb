```python
%pylab inline --no-import-all
import numpy as np
import loewner
plt.style.use('ggplot')
```

```python
# Simulando com driving function linear
t = np.linspace(0,10,1000)
U_linear = t

# Computar o traço da equação de loewner
zchr = loewner.chordal.trace(t, U_linear)  # returns a complex ndarray

```

```python
# Para plotar a função do traço usamos:

fig, axs = plt.subplots(ncols=2, figsize=(10, 5))

axs[0].set_title('Driving Function')
axs[0].plot(t, U_linear, lw=2)

axs[1].set_title('Traço')
axs[1].plot(zchr.real, zchr.imag, lw=2)

axs[0].set_xlabel('$t$', fontsize='xx-large')
axs[0].set_ylabel('$U_t$', fontsize='xx-large')
axs[1].set_aspect('equal')
axs[1].set_xlabel('$Re$', fontsize='xx-large')
axs[1].set_ylabel('$Im$', fontsize='xx-large')
fig.tight_layout()
```
